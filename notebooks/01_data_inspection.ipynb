# Telco Churn Data Inspection

This notebook performs an initial inspection of the raw Telco churn dataset. The goal is to understand the dataset structure, basic quality signals, and target distribution before any modeling work.

## 1. Load the Dataset

Load the raw CSV file into a pandas DataFrame so we can inspect its contents.

```python
from pathlib import Path

import pandas as pd

data_path = Path("../data/raw/telco_churn.csv")
if not data_path.exists():
    data_path = Path("../data/raw/telco_churn.csv.csv")

df = pd.read_csv(data_path)
df
```

## 6. Missing Values

Count missing values per column to identify fields that may need cleaning later.

```python
missing_values = df.isna().sum().sort_values(ascending=False)
missing_values
```

## 7. Churn Distribution

Inspect the target variable counts and percentage distribution to see whether the classes are balanced.

```python
churn_counts = df["Churn"].value_counts(dropna=False)
churn_percentages = df["Churn"].value_counts(normalize=True, dropna=False).mul(100).round(2)

churn_distribution = pd.DataFrame({
    "count": churn_counts,
    "percentage": churn_percentages,
})

churn_distribution
```
